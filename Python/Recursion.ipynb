# Recursion

```python
# Infinite Recursion but in python recursion is terminated if it exceeds 987
def greet():
    print("Amaan")
    greet()

greet()
```

```python
# Head Recursion
count = 0
def greet():
    global count # tells python to use global vairiable
    if(count==4):
        return
    
    count+=1
    print("Amaan")
    greet()

greet()
```

```python
# Tail recursion
count = 0
def greet():
    global count
    if(count==4):
        return
    count += 1
    greet()
    print("Amaan")

greet()
```
